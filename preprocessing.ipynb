```python
import pandas as pd
data=pd.read_csv("Copy of Online Retail.csv")
data.head()
data.info()
data.describe()
```

```python
data = data[~data["InvoiceNo"].str.startswith("C")]
```

cleaning the description coloumn !!

```python
data["Description"].isnull().sum()
data = data[data["Description"].notna()]
data["Description"].isnull().sum()
data["Description"] = data["Description"].str.strip()
data["Description"] = data["Description"].str.lower()
product_counts = data["Description"].value_counts()
data = data[data["Description"].isin(product_counts[product_counts > 5].index)]
keys=["????","discount","postage","manual","adjustment","wet","refund","damages"]
data = data[~data["Description"].isin(keys)]
data.shape
data.head()
```

Cleaning Quantities now!!

```python
data["Quantity"].describe()
data = data[data["Quantity"] > 0] # removed negative and zero quantities

#handle outliers using quantile method 
Q1 = data["Quantity"].quantile(0.25)
Q3 = data["Quantity"].quantile(0.75)

IQR =Q3-Q1
upper_limit =Q3 + 1.5*IQR
data = data[data["Quantity"]<=upper_limit]
```

cleaning the stockcode

```python
data["StockCode"] = data["StockCode"].astype(str)
data["StockCode"] = data["StockCode"].str.upper()
data["StockCode"].str.isalpha().sum()
data = data[~data["StockCode"].str.isalpha()]
data["StockCode"].unique()
data["StockCode"].str.isalpha().sum()
data["StockCode"].dtype
```

Cleaning the invoiceNumber

```python
#invoicedate is cleaned here!
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])
data["InvoiceDate"].dtype
data["InvoiceNo"].isnull().sum()
data["Year"] = data["InvoiceDate"].dt.year
data["Month"] = data["InvoiceDate"].dt.month
data["Day"] = data["InvoiceDate"].dt.day
data["Hour"] = data["InvoiceDate"].dt.hour
data["DayOfWeek"] = data["InvoiceDate"].dt.day_name()
data["InvoiceDate"].dt.year.head()
```

```python
#unitprice is cleaned here!
data["UnitPrice"].dtype
data["UnitPrice"].min()
data["UnitPrice"].describe()
data = data[data["UnitPrice"] > 0] # removed zero and negative unit prices
#i didnt do the outlier treatment for unitprice because the max was reasonable!
```

```python
#for country
data["Country"].unique()
data['Country'].value_counts()
data["Country"] = data["Country"].str.strip()
```

```python
# Save cleaned data to CSV for use by apriori.ipynb
cleaned_path = "cleaned_retail.csv"
data.to_csv(cleaned_path, index=False)
print(f"Cleaned data saved to {cleaned_path}")
```
